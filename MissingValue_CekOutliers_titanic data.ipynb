```python
import pandas as pd 
import numpy as np

df=pd.read_csv('titanic_data.csv')
df
```

```python
#label encoding fungsi untuk mengubah setiap nilai dalam kolom menjadi angka yang berurutan 

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df
```

```python
#isnull digunakan untuk cek missing value atau tidak 

print(df.isnull().sum())
```

```python
# drop digunakan untuk memotong data kolom
df.drop('Cabin', axis=1, inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
#mengisi missing value pada age dengan rata-rata (mean)

rata2_umur=df['Age'].mean() #mencari rata2 pada atribut age 
df['Age']=df['Age'].fillna(rata2_umur) #fillna digunakan untuk mengisi missing value
df
```

```python
print(df.isnull().sum())
```

```python
#mengisi missing value dengan modus

modus=df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(modus)
df
```

```python
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
#Pengecekan Outliers pada atribut Age

import seaborn as sns 
sns.boxplot(x='Age', data=df) 
```

```python
#hitung nilai Q1 dan Q3 pada atribut Age

umur=df['Age']
Q1=np.quantile(umur,.25)
Q3=np.quantile(umur,.75)

print("Kuartil 1 =", Q1)
print("Kuartil 3 =", Q3)
```

```python
#hitung IQR pada atribut Age

IQR= Q3-Q1
print("Inter Quartil =", IQR)
```

```python
#hitung Batas Atas IQR dan Batas Atas IQR pada atribut Age

BB_IQR=Q1-(1.5*IQR)
BA_IQR=Q3+(1.5*IQR)

print("Batas Bawah =",BB_IQR)
print("Batas Atas =",BA_IQR)
```

```python
#Mencari outliers atas bawah pada atribut Age

Outliers_Bawah=[]
Outliers_Atas=[]

for i in umur:
    if(i<BB_IQR):
        Outliers_Bawah.append(i)

    if(i>BA_IQR):
        Outliers_Atas.append(i)

print('Outliers_Bawah =',Outliers_Bawah)
print()
print('Outliers_Atas =',Outliers_Atas)
```

```python
#Mengambil data selain outliers pada atribut Age

data_nonOutliers=df[((df["Age"]>BB_IQR)&(df["Age"]<BA_IQR))]
print("data_nonOutliers pada atribut Age")
print()
data_nonOutliers
```
