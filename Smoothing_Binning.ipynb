```python
import pandas as pd 
import numpy as np

df=pd.read_csv('titanic_data.csv')
df
```

```python
#label encoding fungsi untuk mengubah setiap nilai dalam kolom menjadi angka yang berurutan 

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df
```

```python
#isnull digunakan untuk cek missing value atau tidak 

print(df.isnull().sum())
```

```python
# drop digunakan untuk memotong data kolom
df.drop('Cabin', axis=1, inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
#mengisi missing value pada age dengan rata-rata (mean)

rata2_umur=df['Age'].mean() #mencari rata2 pada atribut age 
df['Age']=df['Age'].fillna(rata2_umur) #fillna digunakan untuk mengisi missing value
df
```

```python
print(df.isnull().sum())
```

```python
#mengisi missing value dengan modus

modus=df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(modus)
df
```

```python
#linspace untuk membuat interval (untuk cut)

bins= np.linspace(min(df['Age']), max (df['Age']), 4)
bins
```

```python
kategori= ['Anak','Muda','Tua']
```

```python
#cut digunakan untuk membagi data menjadi bin
df['Age_binned']=pd.cut(df['Age'],bins, labels=kategori, include_lowest=True )
df
```

```python
kategori=['Murah', 'Sedang', 'Mahal']
df['Fare_binned']=pd.qcut(df['Fare'], 3, labels=kategori)
df

#perbedaan cut dan qcut
#cut membutuhkan linspace sedangkan qcut tidak
```
