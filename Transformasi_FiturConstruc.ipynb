```python
import pandas as pd 
import numpy as np
df=pd.read_csv('titanic_data.csv')
df
```

```python
#label encoding fungsi untuk mengubah setiap nilai dalam kolom menjadi angka yang berurutan 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
#df
```

```python
#isnull digunakan untuk cek missing value atau tidak 
print(df.isnull().sum())
```

```python
# drop digunakan untuk memotong data kolom
df.drop('Cabin', axis=1, inplace=True)
# df
```

```python
df.dtypes
```

```python
#mengisi missing value pada age dengan rata-rata (mean)
rata2_umur=df['Age'].mean() #mencari rata2 pada atribut age 
df['Age']=df['Age'].fillna(rata2_umur) #fillna digunakan untuk mengisi missing value
# df
```

```python
print(df.isnull().sum())
```

```python
#mengisi missing value dengan modus
modus=df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(modus)
# df
```

```python
df['Family_size']=df['SibSp']+df['Parch']+1
df
```

```python
def myfunc(num):
    if num == 1:
        return 'alone'
    elif num>1 and num <=4:
        return 'small_family'
    else :
        return 'large_family'
    
```

```python
df['Family_tipe']=df['Family_size'].apply(myfunc)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Ticket']=le.fit_transform(df['Ticket'])
df
```

```python
#OneHotEncoder (Embarked) = membuat kolom berdasarkan banyak jenis datanya

from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(sparse=False)
y=enc.fit_transform(df[['Embarked']])
y=pd.DataFrame(y)
df.join(y, lsuffix='_caller', rsuffix='_other')
```

```python
enc.categories_
```

```python
#OneHotEncoder 
#Pclass mendefinisikan (1=Upper, 2=Middle, 3=Lower)

from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(sparse=False)
y=enc.fit_transform(df[['Pclass']])
y=pd.DataFrame(y)
df.join(y, lsuffix='_caller', rsuffix='_other') #untuk mengabungkan
```

```python
#OrdianlEncoder 

from sklearn.preprocessing import OrdinalEncoder
kategori=['alone','small_family','large_family']
oce=OrdinalEncoder(categories=[kategori])
df['Family_tipe']=oce.fit_transform(df['Family_tipe'].values.reshape(-1,1))
df
```

```python
def myfunc(num):
    if num == 1:
        return 'Upper'
    elif num == 2:
        return 'Middle'
    else :
        return 'Lower'
    
```

```python
df['Pclass_tipe']=df['Pclass'].apply(myfunc)
df
```

```python
#OrdianlEncoder 

from sklearn.preprocessing import OrdinalEncoder
kategori=['Upper','Middle','Lower']
oce=OrdinalEncoder(categories=[kategori])
df['Pclass_tipe']=oce.fit_transform(df['Pclass_tipe'].values.reshape(-1,1))
df
```
